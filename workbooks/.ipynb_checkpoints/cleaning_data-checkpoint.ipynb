```python
# Check to make sure fighter_df holds both red_df and blue_df
fighter_df.count() == red_df.count() + blue_df.count()
```

```python
fighter_df = fighter_df.sort_values(by=["date"], ascending=False, ignore_index=True)
fighter_df.head(50)
```
